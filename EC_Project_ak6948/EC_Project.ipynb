# Creating Year to Date (YTD) Dataframe

```python
q1["MURDER"] = pd.to_numeric(q1["MURDER"])
q2["MURDER"] = pd.to_numeric(q2["MURDER"])
q3["MURDER"] = pd.to_numeric(q3["MURDER"])
q1["RAPE"] = pd.to_numeric(q1["RAPE"])
q2["RAPE"] = pd.to_numeric(q2["RAPE"])
q3["RAPE"] = pd.to_numeric(q3["RAPE"])
q1["ROBBERY"] = pd.to_numeric(q1["ROBBERY"])
q2["ROBBERY"] = pd.to_numeric(q2["ROBBERY"])
q3["ROBBERY"] = pd.to_numeric(q3["ROBBERY"])
q1["FELONY ASSAULT"] = pd.to_numeric(q1["FELONY ASSAULT"])
q2["FELONY ASSAULT"] = pd.to_numeric(q2["FELONY ASSAULT"])
q3["FELONY ASSAULT"] = pd.to_numeric(q3["FELONY ASSAULT"])
q1["BURGLARY"] = pd.to_numeric(q1["BURGLARY"])
q2["BURGLARY"] = pd.to_numeric(q2["BURGLARY"])
q3["BURGLARY"] = pd.to_numeric(q3["BURGLARY"])
q1["GRAND LARCENY"] = pd.to_numeric(q1["GRAND LARCENY"])
q2["GRAND LARCENY"] = pd.to_numeric(q2["GRAND LARCENY"])
q3["GRAND LARCENY"] = pd.to_numeric(q3["GRAND LARCENY"])
q1["GRAND LARCENY OF MOTOR VEHICLE"] = pd.to_numeric(q1["GRAND LARCENY OF MOTOR VEHICLE"])
q2["GRAND LARCENY OF MOTOR VEHICLE"] = pd.to_numeric(q2["GRAND LARCENY OF MOTOR VEHICLE"])
q3["GRAND LARCENY OF MOTOR VEHICLE"] = pd.to_numeric(q3["GRAND LARCENY OF MOTOR VEHICLE"])
q1["TOTAL"] = pd.to_numeric(q1["TOTAL"])
q2["TOTAL"] = pd.to_numeric(q2["TOTAL"])
q3["TOTAL"] = pd.to_numeric(q3["TOTAL"])
```

```python
ytd = pd.DataFrame()
ytd["Park"] = q3.PARK
ytd["Borough"] = q3.BOROUGH
ytd["Acres"] = q3["SIZE (ACRES)"]
ytd["Murder"] = q1["MURDER"] + q2["MURDER"] + q3["MURDER"]
ytd["Rape"] = q1.RAPE + q2.RAPE + q3.RAPE
ytd["Robbery"] = q1.ROBBERY + q2.ROBBERY + q3.ROBBERY
ytd["Felony Assault"] = q1["FELONY ASSAULT"] + q2["FELONY ASSAULT"] + q3["FELONY ASSAULT"]
ytd["Burglary"] = q1.BURGLARY + q2.BURGLARY + q3.BURGLARY
ytd["Grand Larceny"] = q1["GRAND LARCENY"] + q2["GRAND LARCENY"] + q3["GRAND LARCENY"]
ytd["Grand Larceny of Motor Vehicle"] = q1["GRAND LARCENY OF MOTOR VEHICLE"] \
                                    + q2["GRAND LARCENY OF MOTOR VEHICLE"] \
                                    + q3["GRAND LARCENY OF MOTOR VEHICLE"]
ytd["Total"] = q1.TOTAL + q2.TOTAL + q3.TOTAL
ytd.head(10)
```

## Training and Testing

```python
labels = np.array(parksModel.condition)
```

```python
feature_list = list(parksModel.drop(['Prop ID', 'condition', 'Park', 'Borough', 'Acres'], axis=1))
```

```python
feature_list
```

```python
# Not important
feat_list = list(parksModel.drop(['Prop ID', 'Park', 'Borough', 'Acres'], axis=1))
feat_list
```

```python
features = np.array(parksModel.drop(['Prop ID', 'condition', 'Park', 'Borough', 'Acres'], axis=1))
```

```python
features.shape
```

```python
from sklearn.ensemble  import RandomForestClassifier as rfc
from sklearn.cross_validation import train_test_split
from sklearn.metrics import confusion_matrix

train_features, test_features, train_labels, \
            test_labels = train_test_split(features, labels, 
                                           test_size = 0.25, random_state = 42)
```

```python
test_labels[np.isnan(test_labels)] = 0
```

```python
print('Training Features Shape:', train_features.shape)
print('Training Labels Shape:', train_labels.shape)
print('Testing Features Shape:', test_features.shape)
print('Testing Labels Shape:', test_labels.shape)
```

```python
# Import the model we are using
from sklearn.ensemble import RandomForestRegressor
# Instantiate model with 10000 decision trees
rf = RandomForestRegressor(n_estimators = 10000, random_state = 42)
# Train the model on training data
rf.fit(train_features, train_labels)
```

```python
# Use the forest's predict method on the test data
predictions = rf.predict(test_features)
```

```python
rf.score(test_features, test_labels)
```

```python
predictions
```

```python
# The baseline predictions are the historical averages
baseline_preds = test_features[:, feat_list.index('condition')]
# Baseline errors, and display average baseline error
baseline_errors = abs(baseline_preds - test_labels)
# baseline_errors[np.isnan(baseline_errors)] = 0
print('Average baseline error: ', round(np.mean(baseline_errors), 2))
```

```python
# Calculate the absolute errors
errors = abs(predictions - test_labels)
# errors[np.isnan(errors)] = 0
# Print out the mean absolute error (mae)
print('Mean Absolute Error:', np.mean(errors))
```

```python
# Calculate mean absolute percentage error (MAPE)
mape = 100 * (errors / test_labels)
# mape[np.isnan(mape)] = 0
# Calculate and display accuracy
accuracy = 100 - np.mean(mape)
print('Accuracy:', round(accuracy, 2), '%.')
```

```python
# Get numerical feature importances
importances = list(rf.feature_importances_)
# List of tuples with variable and importance
feature_importances = [(feature, round(importance, 2)) for feature, importance in zip(feature_list, importances)]
# Sort the feature importances by most important first
feature_importances = sorted(feature_importances, key = lambda x: x[1], reverse = True)
# Print out the feature and importances 
[print('Variable: {:20} Importance: {}'.format(*pair)) for pair in feature_importances];
```
