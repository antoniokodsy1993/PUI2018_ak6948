```python
try:
    import urllib2 as ulib
except ImportError:
    import urllib3 as ulib

try:
    from StringIO import BytesIO as io
except ImportError:
    from io import BytesIO as io

try:
    from urllib import urlencode as urlencode
except ImportError:
    from urllib.parse import urlencode as urlencode
    
try:
    from urllib import urlopen as urlopen
except ImportError:
    from urllib.request import urlopen as urlopen
    
try:
    from urllib2 import HTTPError as HTTPError
except ImportError:
    from urllib.error import HTTPError as HTTPError

import ast
import pandas as pd
```

```python
API_URL = 'https://fb55.carto.com:443/api/v2/sql?q='

def queryCartoDB(query, formatting = 'CSV', source = API_URL):
    '''queries carto datasets from a given carto account
    Arguments: 
    query - string: a valid sql query string
    format - outlut format  OPTIONAL (default CSV)
    source - a valid sql api endpoint OPTIONAL (default carto fb55 account)
    Returns:
    the return of the sql query AS A STRING
    NOTES:
    designed for the carto API, tested only with CSV return format'''
    
    
    data = urlencode({'format': formatting, 'q': query}).encode("utf-8")
    try:
        response = urlopen(source, data)
        return response.read()
    except HTTPError as e:
        raise (ValueError('\n'.join(ast.literal_eval(e.readline())['error'])))
        
def get_data(query):
    try:
        return pd.read_csv(io(queryCartoDB(query)), sep = ',')
    except ValueError as v:
        print (str(v))
```

```python
#Showing the top 10 records
query = ''' SELECT * FROM fb55.citibike
            ORDER BY start_station_id ASC, tripduration ASC
            LIMIT 10
            ; '''
table = get_data(query)
table
```

```python
#showing the last 10 records
query = ''' SELECT * FROM fb55.citibike
            ORDER BY start_station_id DESC, tripduration DESC
            LIMIT 10
            ; '''
table = get_data(query)
table
```

```python
#List all unique start_station_id values
query = ''' SELECT DISTINCT start_station_id FROM fb55.citibike
            ORDER BY start_station_id 
            ; '''
table = get_data(query)
table
```

```python
#Number of Trips
query = ''' SELECT COUNT(cartodb_id) FROM fb55.citibike
            ; '''
table = get_data(query)
table
```

```python
#Average, minimum, and maximum trip duration 
query = ''' SELECT AVG(tripduration) as average, MIN(tripduration) as minimum, MAX(tripduration) as maximum FROM fb55.citibike
            
            ; '''
table = get_data(query)
table
```

```python
#Average trip duration for trips started on the weekdays
query = ''' SELECT AVG(tripduration)
            FROM fb55.citibike
            WHERE EXTRACT(dow FROM starttime) NOT IN (0, 6)
            ; '''
table = get_data(query)
table
```

```python
#Average trip duration for trips started on the weekdays
query = ''' SELECT start_station_id, start_station_latitude, start_station_longitude
            FROM fb55.citibike
            GROUP BY start_station_id, start_station_longitude, start_station_latitude
            ORDER BY start_station_id
            ; '''
table = get_data(query)
table
```

```python
#Number of trips started per station
query = ''' SELECT start_station_id, COUNT(start_station_id) as trip_count
            FROM fb55.citibike
            GROUP BY start_station_id
            ORDER BY start_station_id
            ; '''
table = get_data(query)
table
```

```python
#Number of trips started per station for stations within 500m of Time Square
query = ''' SELECT start_station_id, COUNT(start_station_id) as trip_count
            FROM fb55.citibike
            WHERE acos(sin(40.7577*(pi()/180)) * sin(start_station_latitude*(pi()/180)) + cos(40.7577*(pi()/180)) * cos(start_station_latitude*(pi()/180)) * cos(start_station_longitude*(pi()/180) - (-73.9857*(pi()/180)))) * 6371 <= 0.5
            GROUP BY start_station_id
            ORDER BY start_station_id
            ; '''
table = get_data(query)
table
```

```python
#The station that had the longest average trip duration during weekends and within 500m of TimeSquare
query = ''' SELECT start_station_id, AVG(tripduration) as average
            FROM fb55.citibike
            WHERE EXTRACT(dow FROM starttime) IN (0, 6) AND acos(sin(40.7577*(pi()/180)) * sin(start_station_latitude*(pi()/180)) + cos(40.7577*(pi()/180)) * cos(start_station_latitude*(pi()/180)) * cos(start_station_longitude*(pi()/180) - (-73.9857*(pi()/180)))) * 6371 <= 0.5
            GROUP BY start_station_id
            ORDER BY average DESC
            LIMIT(1)
            ; '''
table = get_data(query)
table
```
